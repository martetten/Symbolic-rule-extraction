```python
def measure_intervention_magnitude(model, sae_model, texts, p_lens, lat_idx, h_name, bs=32, dev="cuda"):
    """Возвращает массивы косинусного сходства и L2-нормы разницы для каждого примера."""
    cos_sims = []
    norm_diffs = []
    hook = create_ablation_hook(sae_model, lat_idx, dev)

    with torch.no_grad():
        for i in range(0, len(texts), bs):
            batch_txt = texts[i:i+bs]
            batch_lens = p_lens[i:i+bs]
            toks = model.to_tokens(batch_txt, prepend_bos=True).to(dev)

            # Исходные активации
            _, cache_orig = model.run_with_cache(toks, names_filter=lambda n: n == h_name)

            # Активации после абляции (контекст хуков гарантирует применение интервенции)
            with model.hooks(fwd_hooks=[(h_name, hook)]):
                _, cache_abl = model.run_with_cache(toks, names_filter=lambda n: n == h_name)

            for b in range(toks.shape[0]):
                seq_len = batch_lens[b]
                safe_idx = min(seq_len, cache_orig[h_name].shape[1] - 1)
                orig_vec = cache_orig[h_name][b, safe_idx, :].to(torch.float32)
                abl_vec = cache_abl[h_name][b, safe_idx, :].to(torch.float32)

                cos_sims.append(torch.nn.functional.cosine_similarity(orig_vec, abl_vec, dim=0).item())
                norm_diffs.append(torch.norm(orig_vec - abl_vec).item())

    return np.array(cos_sims), np.array(norm_diffs)
```

```python
test_latent = candidate_latents[0]
```

```python
cos_sims_arr, norm_diffs_arr = measure_intervention_magnitude(
    hooked_model, sae, texts_causal, prompt_lens, test_latent, HOOK_NAME, bs=BATCH_SIZE, dev="cuda"
)
```

```python
print(f"\nРезультаты для латента #{test_latent}:")
print(f"  Среднее косинусное сходство (orig vs ablated): {cos_sims_arr.mean():.4f}")
print(f"  Медиана косинусного сходства:                {np.median(cos_sims_arr):.4f}")
print(f"  Средняя норма разницы (||Δ||):               {norm_diffs_arr.mean():.4f}")
print(f"  Медиана нормы разницы:                       {np.median(norm_diffs_arr):.4f}")
```
